```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('yelp_academic_dataset_user.csv')
```

```python
df.head()
```

```python
print("Number of users :", df.shape[0])
print("Number of attributes per user :", df.shape[1])
```

```python
df["yelping_since"] = df["yelping_since"].astype("datetime64")
pltd = df.groupby(df["yelping_since"].dt.year)['user_id'].count().plot(kind="bar", title='User growth over the years')
pltd.set_xlabel('Year')
pltd.set_ylabel('Number')
```
