*Data is highly skewed as seen in the first histogram* <br>
Log transform is applied to the data <br>
*Log transformed data is plotted using a boxplot and a histogram* <br>

```python
sum_check=df.sum(axis=1)
#plot histogram
plt.hist(sum_check)
plt.show()
#log transform applied to data
a=sum_check.apply(np.log)
#plot boxplot and histogram
plt.boxplot(a)
plt.show()
a.hist()
plt.show()

print(sum_check.value_counts)
```
